### data

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
p_loss = pd.read_csv('p_loss.csv')
plt.figure(figsize=(10, 5))
plt.plot(p_loss.values, label='loss')
plt.plot(p_loss.rolling(5).mean().values, label='5-mean')
plt.plot(p_loss.rolling(10).mean().values, label='10-mean')

plt.legend()
plt.show()
```
